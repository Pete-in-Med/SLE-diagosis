```python
#SLE diagnosis
#First step : classify by ANA
while True :
  #input ANA(antinuclear antibodies) at a titer of ≥1:80 on HEp-2 cells or an equivalent positive test (ever)
  str1 = input("Do the patient has his or her ANA value?(y/n)")
  #ensure that the system wouldn't break owing to uppercase and lowercase problems
  upstr1 = str1.upper()
  if upstr1 == "N":
    print("This patient could not be classify as SLE")
    break
  elif upstr1 =="Y" :
    print("Please read the following descriptions before entering the criteria:")
    print('''1. Do not count a criterion if there is a more likely explanation than SLE
      2. Criteria need not occur simultaneously.''')
    #start to input data
    ANA_value = input("Please enter the value of ANA at a titer of ≥1:80 on HEp-2 cells:")
    break
  else :
    print('Please check to ensure that the input is "y" or "n", and input the data again')
    continue


#input data
#Second step : ask for the condition of the patient
#clinical domains and criteria
list_clinical_domain = ["Fever","Autoimmune hemolysis","Thrombocytopenia","Leukopenia","Seizure","Psychosis","Delirium","Acute cutaneous lupus","Subacute cutaneous OR discoid lupus",
"Oral ulcers","Non-scarring alopecia","Acute pericarditis","Pleural or pericardial effusion","Joint involvement","Renal biopsy Class III or IV lupus nephritis", 
"Renal biopsy Class II or V lupus nephritis", "Proteinuria >0.5g/24h"]
list_immunology_domain = ["Anti-cardiolipin antibodies OR Anti-β2GP1 antibodies OR  Lupus anticoagulant","Low C3 AND low C4","Low C3 OR low C4 ","Anti-dsDNA antibody* OR Anti-Smith antibody"]
list_clinical_result_A = []
list_clinical_result_1 = []
list_immunology_result_B = []
list_immunology_result_2 = []

i = 0
print()
print("The following is the input process of the clinical data!")
while i < len(list_clinical_domain):
  print(list_clinical_domain[i],"?","(y/n)")
  r = input()
  #ensure that the system wouldn't break owing to uppercase and lowercase problems
  up_r = r.upper()
  if up_r == "N":
    list_clinical_result_A.append(up_r)
    up_r = 0 
    list_clinical_result_1.append(up_r) # adding the element
    i += 1
  elif up_r =="Y" :
    list_clinical_result_A.append(up_r)
    up_r = 1
    list_clinical_result_1.append(up_r) # adding the element
    i += 1
  elif up_r =="NI" :
    list_clinical_result_A.append(up_r)
    up_r = 0
    list_clinical_result_1.append(up_r) # adding the element
    i += 1
  else :
    print('Please check to ensure that the input is "y" / "n" / "ni", and input the data again')

i = 0
print() #print an empty space to let the system be easier to read
print("The following is the input process of the immunology data!")
while i < len(list_immunology_domain):
  print(list_immunology_domain[i],"?","(y/n)")
  r = input()
  #ensure that the system wouldn't break owing to uppercase and lowercase problems
  up_r = r.upper()
  if up_r == "N":
    list_immunology_result_B.append(up_r)
    up_r = 0 
    list_immunology_result_2.append(up_r) # adding the element
    i += 1
  elif up_r =="Y" :
    list_immunology_result_B.append(up_r)
    up_r = 1
    list_immunology_result_2.append(up_r) # adding the element
    i += 1
  elif up_r =="NI" :
    list_immunology_result_B.append(up_r)
    up_r = 0
    list_immunology_result_2.append(up_r) # adding the element
    i += 1
  else :
    print('Please check to ensure that the input is "y" / "n" / "ni", and input the data again')

  
#Third step : confirm the data
dic1 = dict(zip(list_clinical_domain, list_clinical_result_A))
dic2 = dict(zip(list_immunology_domain, list_immunology_result_B))
#print readable dictionary
import pprint
print()
print("The following part is about clinical data")
pprint.pprint(dic1)
print()
print("The following part is about immunology data")
pprint.pprint(dic2)
while True :
  confirm = input("Please confirm whether the result is correct or not(y/n):")
  up_confirm = confirm.upper()
  if up_confirm == "Y":
    break
  elif up_confirm == "N":
    #quit the system and ask the doctor to input the data again
    print("If you enter the wrong input, please stop the system and restart the system again")
    break
  else :
    print('Please check to ensure that the input is "y" or "n", and input the data again')
    continue
  

#Forth step : dealing with data
#patient result
#using the list_clinical_result_1 to calculate the total score
list_patient_Constitutional = list_clinical_result_1[0:1]
list_patient_Hematologic = list_clinical_result_1[1:4]
list_patient_Neuropsychiatric = list_clinical_result_1[4:7]
list_patient_Mucocutaneous = list_clinical_result_1[7:11]
list_patient_Serosal = list_clinical_result_1[11:13]
list_patient_Musculoskeletal = list_clinical_result_1[13:14]
list_patient_Renal = list_clinical_result_1[14:17]

#using the list_clinical_result_2 to calculate the total score
list_patient_Antiphospholipid_antibodies = list_immunology_result_2[0:1]
list_patient_Complement_proteins = list_immunology_result_2[1:3]
list_patient_SLE_specific_antibodies = list_immunology_result_2[3:4]

#weight
list_weight_Constitutional = [2]
list_weight_Hematologic = [4,4,3]
list_weight_Neuropsychiatric = [5,3,2]
list_weight_Mucocutaneous = [6,4,2,2]
list_weight_Serosal = [6,5]
list_weight_Musculoskeletal = [6]
list_weight_Renal = [10,8,4]
list_weight_Antiphospholipid_antibodies = [2]
list_weight_Complement_proteins = [4,3]
list_weight_SLE_specific_antibodies = [6]

#value
list_patient = [list_patient_Constitutional,list_patient_Hematologic,list_patient_Neuropsychiatric,list_patient_Mucocutaneous,list_patient_Serosal,list_patient_Musculoskeletal,
                list_patient_Renal,list_patient_Antiphospholipid_antibodies,list_patient_Complement_proteins,list_patient_SLE_specific_antibodies]
list_weight = [list_weight_Constitutional,list_weight_Hematologic,list_weight_Neuropsychiatric,list_weight_Mucocutaneous,list_weight_Serosal,list_weight_Musculoskeletal,
               list_weight_Renal,list_weight_Antiphospholipid_antibodies,list_weight_Complement_proteins,list_weight_SLE_specific_antibodies]
list_value = []
for i in range(0,len(list_patient)) :
  value = max([x*y for x,y in zip(list_patient[i],list_weight[i])])
  list_value.append(value)

#main cause

#sum
sum_up = 0
sum_up = sum(list_value)
print(sum_up)

#output
#classification
if sum_up >= 10 :
  print("The patient is diagnosed as SLE")
else :
  print("The current test project does not provide sufficient evidence to determine a positive result")

#all No Information columns in the list
print("If you need more precise results, you need to complete the patient's data by doing the following tests")
print("Here are the missing data of the patient in the clinical domain")
for key,value in dic1.items():  
    if value == "NI":
        print(key)
print("Here are the missing data of the patient in the immunology domain")
for key,value in dic2.items():  
    if value == "NI":
        print(key) 

#distinguish the main cause of the disease
list_all_domain = list_clinical_domain + list_immunology_domain
dic3 = dict(zip(list_all_domain,list_value))
max_key = max(dic3, key=dic3.get)
print("Here is the main cause of the patient to be diagnosed as an SLE patient")
print(max_key)
```
